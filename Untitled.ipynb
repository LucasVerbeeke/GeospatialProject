```python
# By : Ayman Mutasim

import numpy as np
import matplotlib.pyplot as plt
import rasterio

# --- CONFIGURATION & DATA PATHS ---
RASTER_FILE_PATH = "Data/Classified_Clipped_LC_Belgium_2022.tif"
MAX_CLUSTERS_TO_TEST = 10
OUTPUT_RASTER_PATH = "OUTPUT FILE PATH AND NAME"

# --------------------------------------------------------
# EXTRA FUNCTION: PRINT FLOAT CENTROIDS
# --------------------------------------------------------

def show_cluster_centroids(centroids):
    """
    Prints the FLOAT centroid values produced by K-Means.
    """
    print("\nðŸŽ¯ FINAL FLOAT CENTROIDS FROM K-MEANS:")
    print(np.array(centroids))
    print(f"\nðŸ“Œ Number of centroids: {len(centroids)}")


# --------------------------------------------------------
# 4. CORE K-MEANS LOGIC (Single Band)
# --------------------------------------------------------

def kmeans_single_band(data_2d, k, max_iters=100, tolerance=1e-4, seed=42):
    """Runs K-Means on a 2D raster array and returns:
       - cluster labels,
       - FLOAT centroids,
       - SSD value.
    """
    x, y = data_2d.shape
    reshaped_data = data_2d.reshape(-1, 1)

    np.random.seed(seed)
    centroids = reshaped_data[np.random.choice(reshaped_data.shape[0], k, replace=False)].flatten().astype(np.float64)

    for _ in range(max_iters):
        distances = np.abs(reshaped_data - centroids)
        closest_centroids = np.argmin(distances, axis=1)
        new_centroids = np.zeros(k, dtype=np.float64)

        for i in range(k):
            cluster_points = reshaped_data[closest_centroids == i]
            if cluster_points.size > 0:
                new_centroids[i] = cluster_points.mean()  # keep as float
            else:
                new_centroids[i] = centroids[i]

        if np.all(np.abs(new_centroids - centroids) < tolerance):
            break

        centroids = new_centroids

    final_distances = np.min(np.abs(reshaped_data - centroids), axis=1)
    ssd = np.sum(final_distances ** 2)

    return closest_centroids.reshape(x, y), centroids, ssd



# --------------------------------------------------------
# 6. SAVE OUTPUT RASTER
# --------------------------------------------------------

def save_classified_raster(output_data, output_path, src_profile):
    """ Writes the classified NumPy array to a GeoTIFF file. """
    print(f"\n--- Saving Classified Raster to: {output_path} ---")

    out_profile = src_profile.copy()
    out_profile.update({
        'dtype': output_data.dtype,
        'count': 1,
        'nodata': None
    })

    try:
        with rasterio.open(output_path, 'w', **out_profile) as dst:
            dst.write(output_data, 1)
        print("Raster successfully saved!")
    except Exception as e:
        print(f"ERROR saving raster: {e}")


# --------------------------------------------------------
# 7. MAIN CLASSIFICATION WORKFLOW
# --------------------------------------------------------

def classify_single_band_raster(file_path, max_k=10, output_path=None):

    # Load raster
    with rasterio.open(file_path) as src:
        if src.count != 1:
            print("âŒ Raster must have exactly 1 band.")
            return

        src_profile = src.profile
        data = src.read(1).astype(np.float64)
        valid_data = data.flatten()

    print(f"Data Loaded: {file_path}. Shape: {data.shape}")

    # --- STEP 2: ELBOW METHOD (IMPROVED PLATEAU VERSION) ---
    ssd_scores = []
    k_range = range(1, max_k + 1)

    print("\n--- Running Elbow Method to find optimal K ---")
    for k in k_range:
        if k == 1:
            mean_val = np.mean(valid_data)
            ssd = np.sum((valid_data - mean_val) ** 2)
        else:
            _, _, ssd = kmeans_single_band(data.copy(), k)

        ssd_scores.append(ssd)
        print(f"Calculated SSD for K={k}: {ssd:.4f}")

    # Convert to numpy for handling
    ssd_scores_arr = np.array(ssd_scores, dtype=float)
    k_values = np.array(list(k_range))

    # Compute relative improvements
    with np.errstate(divide='ignore', invalid='ignore'):
        improvements = (ssd_scores_arr[:-1] - ssd_scores_arr[1:]) / ssd_scores_arr[:-1]
        improvements = np.nan_to_num(improvements, nan=0.0, posinf=0.0, neginf=0.0)

    # If improvement < threshold â†’ considered flat
    IMPROVEMENT_THRESHOLD = 0.05  # 5%
    flat_indices = np.where(improvements < IMPROVEMENT_THRESHOLD)[0]

    if flat_indices.size > 0:
        optimal_k = int(k_values[flat_indices[0] + 1])  # first K where curve flattens
    else:
        optimal_k = max_k

    print(f"\nCalculated Optimal K (Plateau Heuristic): K = {optimal_k}")

    # Plot elbow curve
    plt.figure(figsize=(8, 5))
    plt.plot(k_values, ssd_scores, marker='o', linestyle='--')
    plt.plot(optimal_k, ssd_scores[optimal_k - 1], "ro", markersize=10, label=f"Optimal K = {optimal_k}")
    plt.title("Elbow Method (Plateau-Based)")
    plt.xlabel("Number of Clusters (K)")
    plt.ylabel("Sum of Squared Distances (SSD)")
    plt.grid()
    plt.legend()
    plt.show()

    # --- STEP 3: FINAL CLASSIFICATION ---
    segmented_labels, final_centroids, _ = kmeans_single_band(data, optimal_k)

    print("\n--- Final Classification Results ---")
    print(f"Final FLOAT Centroids for K={optimal_k}: {final_centroids}")

    show_cluster_centroids(final_centroids)

    # Map cluster labels â†’ FLOAT centroid values
    out_data_flat = np.zeros(data.size, dtype=np.float64)
    flat_labels = segmented_labels.flatten()

    for i in range(optimal_k):
        out_data_flat[flat_labels == i] = final_centroids[i]

    classified_raster = out_data_flat.reshape(data.shape)

    # Save raster
    if output_path:
        save_classified_raster(classified_raster, output_path, src_profile)

    # Plot Label Image
    plt.figure(figsize=(10, 8))
    plt.imshow(segmented_labels, cmap="nipy_spectral")
    plt.colorbar(label="Cluster Label")
    plt.title(f"K-Means Segmentation (K={optimal_k})")
    plt.show()


# ---------------------------------------------------------
# RUN WORKFLOW
# ---------------------------------------------------------

classify_single_band_raster(RASTER_FILE_PATH, MAX_CLUSTERS_TO_TEST, OUTPUT_RASTER_PATH)
```
